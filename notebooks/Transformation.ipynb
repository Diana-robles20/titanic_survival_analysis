```python
import pandas as pd

df = pd.read_csv('../data/processed/titanic_selected.csv')
df.head()
```

**Revisión de valores nulos**

```python
df.isnull().sum()
```

Se decide rellenar ese único valor nulo en la variable Fare con la mediana y no con la media, ya que con base en su distribución (histograma) se observa que hay tarifas relativamente altas, lo cual distorsiona la media , no siendo una representación adecuada de las tarifas

```python
df['fare'] = df['fare'].fillna(df['fare'].median())
```

Se rellenan los valores nulos en la variable 'age' con la mediana. En contraste con la variable fare, la edad presenta una distribución aproximadamente normal, hay una diferencia relativamente baja (1.8) entre la media y la mediana, por lo que ambas son buenas representaciones del conjunto de datos. 

```python
df['age'] = df['age'].fillna(df['age'].median())
```

**Codificación de variables numéricas**

```python
df = pd.get_dummies(df, columns=['sex','embarked'], drop_first=True, dtype = int)
df
```

```python
df.to_csv('../data/processed/titanic_transformation.csv', index=False)
```
